# **Exploring PM2.5 and PM10 air quality trends in New York–Newark–Jersey City (NY-NJ Metropolitan Area).**

# **Introduction**

This project presents a data analysis of air quality in New York–Newark–Jersey City (NY-NJ Metropolitan Area), focusing on the atmospheric pollutants PM2.5 and PM10. These pollutants are particulate matter suspended in the air: PM2.5 refers to fine particles with a diameter of 2.5 micrometers or smaller, while PM10 includes particles up to 10 micrometers in diameter. Due to their small size, these particles can be inhaled and may affect human health, especially the respiratory system.

The objective is to study their evolution over time and identify patterns that provide a better understanding of pollution dynamics in the city.

The work is developed in Python, using specialized libraries for data manipulation, visualization, and analysis. Through this project, techniques such as data cleaning, graphical representation, and trend exploration are applied to obtain clear and structured insights into air quality.

This analysis not only illustrates the historical evolution of these pollutants but also demonstrates how data science tools can be applied to real-world problems with relevance to public health and the environment.

```python
# @title
import matplotlib.pyplot as plt
from pathlib import Path

def plot_roadmap(
    steps,
    orientation="horizontal",  # "horizontal" or "vertical"
    node_size=1000,
    text_rotation=0,
    title="Project Roadmap",
    spacing=1.0,               # distance between nodes
    annotate_above=True        # place labels above (horizontal) or right (vertical)
):
    """
    Plot a simple roadmap diagram.

    Parameters
    ----------
    steps : list[str]
        Labels for each step in order.
    orientation : str
        "horizontal" or "vertical".
    node_size : int or float
        Scatter size for nodes.
    text_rotation : float
        Rotation angle for node labels (degrees).
    title : str
        Figure title.
    spacing : float
        Gap between nodes.
    annotate_above : bool
        If horizontal, True=labels above timeline, False=below.
        If vertical, True=labels to the right, False=to the left.
    """
    if not steps:
        raise ValueError("Provide at least one step")

    fig, ax = plt.subplots(figsize=(10, 3) if orientation=="horizontal" else (6, 6))

    if orientation == "horizontal":
        x_positions = [i * spacing for i in range(len(steps))]
        y_positions = [0] * len(steps)
        for i, (label, x, y) in enumerate(zip(steps, x_positions, y_positions)):
            ax.scatter(x, y, s=node_size, zorder=3)
            y_text = 0.12 if annotate_above else -0.12
            ax.text(x, y + y_text, label, ha="center", va="bottom" if annotate_above else "top",
                    fontsize=9, weight="bold", rotation=text_rotation)
            if i < len(steps) - 1:
                ax.plot([x, x_positions[i+1]], [y, y], linewidth=1.5, zorder=2)
        ax.set_ylim(-0.5, 1.0)
        ax.set_xlim(min(x_positions) - 0.5, max(x_positions) + 0.5)
    elif orientation == "vertical":
        y_positions = [i * spacing for i in range(len(steps))][::-1]
        x_positions = [0] * len(steps)
        for i, (label, x, y) in enumerate(zip(steps, x_positions, y_positions)):
            ax.scatter(x, y, s=node_size, zorder=3)
            x_text = 0.12 if annotate_above else -0.12
            ax.text(x + x_text, y, label,
                    ha="left" if annotate_above else "right",
                    va="center", fontsize=9, weight="bold", rotation=text_rotation)
            if i < len(steps) - 1:
                ax.plot([x, x], [y, y_positions[i+1]], linewidth=1.5, zorder=2)
        ax.set_xlim(-1.0, 1.0)
        ax.set_ylim(min(y_positions) - 0.5, max(y_positions) + 0.5)
    else:
        raise ValueError("orientation must be 'horizontal' or 'vertical'")

    ax.axis("off")
    plt.title(title, fontsize=14, weight="bold")
    plt.tight_layout()
    return fig, ax

# Example usage
example_steps = [
    "Data Collection",
    "Data Cleaning & Preprocessing",
    "Exploratory Data Analysis",
    "Visualization of Trends",
    "Key Findings & Conclusions"
]

fig, ax = plot_roadmap(
    example_steps,
    orientation="horizontal",
    node_size=1000,
    text_rotation=10,
    title="Project Roadmap (Editable Template)",
    spacing=1.2,
    annotate_above=True
)

# Save example image
out_dir = Path("/mnt/data")
out_dir.mkdir(parents=True, exist_ok=True)
png_path = out_dir / "roadmap_timeline.png"
fig.savefig(png_path, dpi=200, bbox_inches="tight")

# Save reusable script
script_code = r'''import matplotlib.pyplot as plt

def plot_roadmap(
    steps,
    orientation="horizontal",
    node_size=1000,
    text_rotation=0,
    title="Project Roadmap",
    spacing=1.0,
    annotate_above=True
):
    if not steps:
        raise ValueError("Provide at least one step")

    fig, ax = plt.subplots(figsize=(10, 3) if orientation=="horizontal" else (6, 6))

    if orientation == "horizontal":
        x_positions = [i * spacing for i in range(len(steps))]
        y_positions = [0] * len(steps)
        for i, (label, x, y) in enumerate(zip(steps, x_positions, y_positions)):
            ax.scatter(x, y, s=node_size, zorder=3)
            y_text = 0.12 if annotate_above else -0.12
            ax.text(x, y + y_text, label, ha="center", va="bottom" if annotate_above else "top",
                    fontsize=9, weight="bold", rotation=text_rotation)
            if i < len(steps) - 1:
                ax.plot([x, x_positions[i+1]], [y, y], linewidth=1.5, zorder=2)
        ax.set_ylim(-0.5, 1.0)
        ax.set_xlim(min(x_positions) - 0.5, max(x_positions) + 0.5)
    elif orientation == "vertical":
        y_positions = [i * spacing for i in range(len(steps))][::-1]
        x_positions = [0] * len(steps)
        for i, (label, x, y) in enumerate(zip(steps, x_positions, y_positions)):
            ax.scatter(x, y, s=node_size, zorder=3)
            x_text = 0.12 if annotate_above else -0.12
            ax.text(x + x_text, y, label,
                    ha="left" if annotate_above else "right",
                    va="center", fontsize=9, weight="bold", rotation=text_rotation)
            if i < len(steps) - 1:
                ax.plot([x, x], [y, y_positions[i+1]], linewidth=1.5, zorder=2)
        ax.set_xlim(-1.0, 1.0)
        ax.set_ylim(min(y_positions) - 0.5, max(y_positions) + 0.5)
    else:
        raise ValueError("orientation must be 'horizontal' or 'vertical'")

    ax.axis("off")
    plt.title(title, fontsize=14, weight="bold")
    plt.tight_layout()
    return fig, ax

if __name__ == "__main__":
    steps = [
        "Data Collection",
        "Data Cleaning & Preprocessing",
        "Exploratory Data Analysis",
        "Visualization of Trends",
        "Key Findings & Conclusions"
    ]
    fig, ax = plot_roadmap(steps, orientation="horizontal", node_size=1000, text_rotation=10,
                           title="Project Roadmap (Editable Template)", spacing=1.2, annotate_above=True)
    fig.savefig("roadmap_timeline.png", dpi=200, bbox_inches="tight")
'''

script_path = out_dir / "roadmap_figure.py"
with open(script_path, "w", encoding="utf-8") as f:
    f.write(script_code)

plt.show()

png_path, script_path
```

```python

# En Google Colab los archivos subidos se guardan en /content
DATA_DIR = Path("/content")

# Rango de años
years = range(2020, 2026)

# Diccionarios para guardar los DataFrames
PM25_datasets = {}
PM10_datasets = {}

# --- Carga PM2.5 ---
print("Cargando datasets PM2.5:")
for y in years:
    fname = f"pm 2.5_{y}_NewYork.csv"
    fpath = DATA_DIR / fname
    if fpath.exists():
        PM25_datasets[y] = pd.read_csv(fpath)
        print(f"  ✔ {fname} ({len(PM25_datasets[y])} filas)")
    else:
        print(f"  ⚠ No encontrado: {fname}")

# --- Carga PM10 ---
print("\nCargando datasets PM10:")
for y in years:
    fname = f"pm 10_{y}_NewYork.csv"
    fpath = DATA_DIR / fname
    if fpath.exists():
        PM10_datasets[y] = pd.read_csv(fpath)
        print(f"  ✔ {fname} ({len(PM10_datasets[y])} filas)")
    else:
        print(f"  ⚠ No encontrado: {fname}")
```

```python
#1) Unificar y limpiar PM2.5 y PM10

# Función para normalizar columnas de fecha y valor
def normalize_df(df, value_candidates, date_candidates=("date_local","Date")):
    # Detectar columnas
    date_col = next((c for c in date_candidates if c in df.columns), None)
    val_col  = next((c for c in value_candidates if c in df.columns), None)
    if date_col is None or val_col is None:
        return pd.DataFrame(columns=["date","value"])
    # Convertir fecha y limpiar
    out = df[[date_col, val_col]].copy()
    out[date_col] = pd.to_datetime(out[date_col], errors="coerce")
    out = out.dropna(subset=[date_col, val_col])
    out = out.rename(columns={date_col: "date", val_col: "value"})
    out = out.sort_values("date")
    return out

# Combinar años para cada contaminante
def combine_years(datasets_dict, value_candidates):
    parts = []
    for y, dfi in sorted(datasets_dict.items()):
        norm = normalize_df(dfi, value_candidates=value_candidates)
        if not norm.empty:
            norm["year"] = y
            parts.append(norm)
    return pd.concat(parts, ignore_index=True) if parts else pd.DataFrame(columns=["date","value","year"])

# Unificados
pm25_all = combine_years(PM25_datasets, value_candidates=("arithmetic_mean","Daily Mean PM2.5 Concentration"))
pm10_all = combine_years(PM10_datasets, value_candidates=("arithmetic_mean","Daily Mean PM10 Concentration"))

print("PM2.5 filas:", len(pm25_all), " rango:", pm25_all["date"].min(), "→", pm25_all["date"].max())
print("PM10  filas:", len(pm10_all), " rango:", pm10_all["date"].min(), "→", pm10_all["date"].max())
```

```python
#2) Agregaciones diarias y mensuales

def daily_with_rolling(df, window=7):
    if df.empty:
        return df
    s = (df.groupby("date", as_index=False)["value"].mean()
           .rename(columns={"value":"daily"}))
    s["roll"] = s["daily"].rolling(window, min_periods=1).mean()
    return s

def monthly_mean(df):
    if df.empty:
        return df
    m = df.copy()
    m["year"]  = m["date"].dt.year
    m["month"] = m["date"].dt.month
    m["month_abbr"] = m["date"].dt.strftime("%b")
    m = (m.groupby(["year","month","month_abbr"], as_index=False)["value"]
           .mean()
           .rename(columns={"value":"mean"}))
    return m.sort_values(["year","month"])

pm25_daily = daily_with_rolling(pm25_all, window=7)
pm10_daily = daily_with_rolling(pm10_all, window=7)

pm25_monthly = monthly_mean(pm25_all)
pm10_monthly = monthly_mean(pm10_all)
```

```python
# 3) Gráficas básicas con matplotlib

import matplotlib.pyplot as plt

def plot_monthly_lines(m, title):
    if m.empty:
        print("No hay datos para graficar."); return
    plt.figure(figsize=(12,6))
    for y, part in m.groupby("year"):
        part = part.sort_values("month")
        plt.plot(part["month"], part["mean"], marker="o", label=str(y))
    plt.xticks(range(1,13), ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"])
    plt.title(title); plt.xlabel("Mes"); plt.ylabel("µg/m³")
    plt.legend(); plt.tight_layout(); plt.show()


def plot_daily(series, title):
    if series.empty:
        print("No hay datos para graficar.")
        return
    plt.figure(figsize=(12,5))
    plt.plot(series["date"], series["daily"], label="Diario")
    plt.plot(series["date"], series["roll"],  label="Media 7 días", linewidth=2)
    plt.title(title)
    plt.xlabel("Fecha")
    plt.ylabel("µg/m³")
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_monthly_lines(pm25_monthly, "PM2.5 promedio mensual por año en NY NJ")
plot_monthly_lines(pm10_monthly, "PM10 promedio mensual por año en NY NJ")

plot_daily(pm25_daily, "PM2.5 diario y media 7 días")
plot_daily(pm10_daily, "PM10 diario y media 7 días")
```

```python
# 4) Comparación y validación rápida

# Comparación de medias móviles
cmp = pd.merge(
    pm25_daily[["date","roll"]].rename(columns={"roll":"pm25_7d"}),
    pm10_daily[["date","roll"]].rename(columns={"roll":"pm10_7d"}),
    on="date", how="inner"
)

plt.figure(figsize=(12,5))
plt.plot(cmp["date"], cmp["pm25_7d"], label="PM2.5 (7 días)", linewidth=2)
plt.plot(cmp["date"], cmp["pm10_7d"], label="PM10  (7 días)", linewidth=2)
plt.title("Comparación PM2.5 vs PM10, medias móviles 7 días")
plt.xlabel("Fecha")
plt.ylabel("µg/m³")
plt.legend()
plt.tight_layout()
plt.show()

# Correlación simple
if not cmp.empty:
    corr = cmp[["pm25_7d","pm10_7d"]].corr().iloc[0,1]
    print(f"Correlación PM2.5 vs PM10 (7 días): {corr:.3f}")
```

```python
#Comparativas PM2.5 vs PM10
#A) Comparativa por año elegido (mismas 12 posiciones de mes)

YEAR = 2025  # cambia el año si quieres

p25_y = pm25_monthly[pm25_monthly["year"] == YEAR].sort_values("month")
p10_y = pm10_monthly[pm10_monthly["year"] == YEAR].sort_values("month")

plt.figure(figsize=(12,6))
plt.plot(p25_y["month"], p25_y["mean"], marker="o", label=f"PM2.5 {YEAR}")
plt.plot(p10_y["month"], p10_y["mean"], marker="o", label=f"PM10 {YEAR}")
plt.xticks(range(1,13), ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"])
plt.title(f"Comparación mensual PM2.5 vs PM10 en {YEAR} (NY–NJ)")
plt.xlabel("Mes"); plt.ylabel("µg/m³"); plt.legend(); plt.tight_layout(); plt.show()
```

```python
#B) Comparativa “estacional” promedio de todos los años

def seasonal_profile(m):
    # promedio por número de mes, agregando todos los años
    s = m.groupby("month", as_index=False)["mean"].mean()
    return s.sort_values("month")

p25_season = seasonal_profile(pm25_monthly)
p10_season = seasonal_profile(pm10_monthly)

plt.figure(figsize=(12,6))
plt.plot(p25_season["month"], p25_season["mean"], marker="o", label="PM2.5 (promedio años)")
plt.plot(p10_season["month"], p10_season["mean"], marker="o", label="PM10 (promedio años)")
plt.xticks(range(1,13), ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"])
plt.title("Comparación estacional PM2.5 vs PM10 (promedio de todos los años)")
plt.xlabel("Mes"); plt.ylabel("µg/m³"); plt.legend(); plt.tight_layout(); plt.show()
```

```python
#4) Relación entre PM2.5 y PM10
import pandas as pd
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool, Select, Div
from bokeh.layouts import column, row


# 1) Unión diaria y columnas de año/mes
cmp = pd.merge(
    pm25_daily[["date","daily"]].rename(columns={"daily":"pm25"}),
    pm10_daily[["date","daily"]].rename(columns={"daily":"pm10"}),
    on="date", how="inner"
).dropna().copy()

cmp["year"] = pd.to_datetime(cmp["date"]).dt.year
cmp["month"] = pd.to_datetime(cmp["date"]).dt.month
MONTH_LABELS = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
cmp["month_label"] = cmp["month"].apply(lambda m: MONTH_LABELS[m-1])

# 2) Widgets
years = sorted(cmp["year"].unique().tolist())
year_select = Select(title="Año", value="Todos", options=["Todos"] + [str(y) for y in years])
month_select = Select(title="Mes", value="Todos", options=["Todos"] + MONTH_LABELS)

# 3) Función para filtrar
def filtered_df():
    df = cmp.copy()
    if year_select.value != "Todos":
        df = df[df["year"] == int(year_select.value)]
    if month_select.value != "Todos":
        df = df[df["month_label"] == month_select.value]
    return df

df0 = filtered_df()

pts_src  = ColumnDataSource({"pm25": df0["pm25"], "pm10": df0["pm10"], "date": pd.to_datetime(df0["date"])})
line_src = ColumnDataSource({"x": [], "y": []})

# 4) Títulos dinámicos
main_title = Div(text="<h2>Relación entre PM2.5 y PM10 (NY–NJ)</h2>")
sub_title  = Div(text="", styles={"font-size": "15px", "color": "gray"})

# 5) Figura
p = figure(
    width=800, height=500,
    x_axis_label="PM2.5 (µg/m³)",
    y_axis_label="PM10 (µg/m³)"
)

p.scatter("pm25", "pm10", source=pts_src, size=6, alpha=0.6, color="navy")
p.line("x", "y", source=line_src, line_width=2, color="firebrick")

p.add_tools(HoverTool(tooltips=[
    ("Fecha", "@date{%F}"),
    ("PM2.5", "@pm25{0.0}"),
    ("PM10",  "@pm10{0.0}")
], formatters={"@date": "datetime"}))

p.toolbar.autohide = True

# 6) Función de actualización
def update(attr, old, new):
    df = filtered_df()
    pts_src.data = {
        "pm25": df["pm25"],
        "pm10": df["pm10"],
        "date": pd.to_datetime(df["date"])
    }
    if len(df) >= 2:
        corr = np.corrcoef(df["pm25"], df["pm10"])[0,1]
        a, b = np.polyfit(df["pm25"], df["pm10"], 1)
        x_line = np.linspace(df["pm25"].min(), df["pm25"].max(), 100)
        y_line = a * x_line + b
        line_src.data = {"x": x_line, "y": y_line}
        sub_title.text = f"<b>Correlación:</b> {corr:.2f} &nbsp;&nbsp; <b>Ecuación:</b> y = {a:.2f}·x + {b:.2f}"
    else:
        line_src.data = {"x": [], "y": []}
        sub_title.text = "<b>Correlación:</b> N/A"

update(None, None, None)

year_select.on_change("value", update)
month_select.on_change("value", update)

show(column(main_title, row(year_select, month_select), sub_title, p))


```
